```python
gn_iter = pd.read_csv("green_tripdata_2019-01.csv.gz.1",compression='gzip', iterator=True, chunksize=1000)
```

```python
gn = next(gn_iter)
```

```python
%time gn.to_sql(name='green_trip_data', con=engine, if_exists='append')
```

```python
 from time import time
```

```python
while True:
    t_start = time()
    gn = next(gn_iter)
    gn.lpep_pickup_datetime = pd.to_datetime(gn.lpep_pickup_datetime)
    gn.lpep_dropoff_datetime = pd.to_datetime(gn.lpep_dropoff_datetime)
    
    gn.to_sql(name='green_trip_data', con=engine, if_exists='append')
    
    t_end = time()
    
    print('Inserted another chunk...,  took %.3f second' %(t_end - t_start))
    
```
